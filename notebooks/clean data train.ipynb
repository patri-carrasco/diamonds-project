Comprobamos los valores únicos de la columna cut, y comprobamos que los valores son string y los queremos convertir en númericas por eso le damos un valor númerico.

```python
df.cut.unique()
```

```python
cut_map = {
    'Ideal': 5,
    'Premium': 4,
    'Very Good': 3,
    'Good': 2,
    'Fair': 1
}
```

```python
cut_map
```

```python
df["cut"] = df.cut.map(cut_map)
df.head()
```

Lo mismo lo hacemos con el color, https://www.bluenile.com/es/education/diamonds/colour

```python
df.color.unique()
```

```python
color_map = { 'D': 6,
             'E':5,
             'G':4,
             'F':3,
             'H':2,
             'I':1,
             'J':0
}
df['color'] = df['color'].map(color_map)
df.head()
```

```python
df.clarity.unique()
```

Repetimos el procedimiento con clarity. 

```python
clarity_map = {'I1':0, 'SI2':1, 'SI1':2, 'VS2':3, 'VS1':4, 'VVS2':5, 'VVS1':6, 'IF':7}

df['clarity'] = df['clarity'].map(clarity_map)
```
